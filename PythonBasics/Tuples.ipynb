#### Tuples

Tuples are used to store multiple items in a single variable.

A tuple is a collection which is ordered and unchangeable.

```python
thistuple = ("apple", "banana", "cherry")
print(thistuple)
type(thistuple)
```

tuple() -> constructor

```python
thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(thistuple)
```

to change the values in tuples, convert it into list then backt to tuple

```python
thistuple = ('apple','banan', 'cat')
print(thistuple)

listtuple = list(thistuple)
print(listtuple)

listtuple.append('dog')

thistuple = tuple(listtuple)
print(thistuple)
```

Add tuple to a tuple

```python
thistuple2 = ('gift',)

print(thistuple+thistuple2)
print(thistuple)
```

Unpack the tuples

```python
tuple_t = tuple(("apple","banana","cat","apple"))
print(tuple_t)

a,b,c,d = tuple_t

print(a,b,c,d)
```

```python
print(tuple_t.index('banana'))
tuple_t.count(2)
```
