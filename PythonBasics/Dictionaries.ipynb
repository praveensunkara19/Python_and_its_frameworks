#### Dictionary
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

```
dic.update(i:"a")
dic.get(i)
dic.pop(i)
dic.popitem(i) -> pop last item

````python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)

print(thisdict["brand"])
````

````python
# Duplicate values will overwrite existing values:

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
````

Dictionary Items - Data Types
* The values in dictionary items can be of any data type:

````python
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}

colors = thisdict["colors"]
colors
````

items() represent the dictionary as object

````python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.items()

print(x) #before the change

car["year"] = 2020

print(x) #after the change
````

Remove

````python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
````

````python
# The popitem() method removes the last inserted item (in versions before 3.7, a random item is removed instead):

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
````

copy dictionaries

````python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = dict(thisdict)
print(mydict)
````

Nested dictionaries

````python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
myfamily
````

````python
myfamily["child1"]
````

````python
for k,v in myfamily.items(): 
    print(k,v)               #keys:values

    for k_sub, v_sub in v.items():
        print(k_sub, v_sub)
````


Maps

````python
dic = {}

fruits = [1,2,3,2,2]

for i in fruits:
    if not dic.get(i):
        dic.update({i:1})
    else:
        dic.update({i:dic.get(i)+1})

dic
````

````python
ls = [1,1,3,2,3,4]

dic = {}

for n in ls:
    dic[n] = dic.get(n, 0) + 1
````

Sorting with in the Dict

````python
dic = {4: 1, 1: 1, -1: 2, 2: 2, 3: 1}

dic.items()
````

````python
# Sorting by basing values
sorted(dic.items(), key=lambda x:x[1], reverse=True)
````

````python
# sorting by basing the keys
sorted(dic.items(), key=lambda x:x[0], reverse=True)
````
