````python
def num(n):
    if n == 0:
        return
    print(n, end=" ")
    return num(n-1)
    
    
print(num(10))
````

````python
def fact(n):
    if n==1 or n == 0:
        return 1
    return n*fact(n-1)

fact(5)
````

````python
def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)

    
fib(7)
````

finding sum with recursion

````python
def sum_list(numbers):
  if len(numbers) == 0:
    return 0
  else:
    return numbers[0] + sum_list(numbers[1:])

my_list = [1, 2, 3, 4, 5]
print(sum_list(my_list))
````

finding max num

````python
def find_max(numbers):
  if len(numbers) == 1:
    return numbers[0]
  else:
    max_of_rest = find_max(numbers[1:])
    return numbers[0] if numbers[0] > max_of_rest else max_of_rest

my_list = [3, 7, 2, 9, 1]
print(find_max(my_list))
````
