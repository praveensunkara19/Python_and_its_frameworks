#### Map

The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.
```
map(function, iterables)


````python
ls = [1,2,3,4]

sq = map(lambda x:x**2, ls)  #returns the map object
type(sq)

print(list(sq))
````

````python
def myfunc(n):
    return n-10


ls = [i for i in range(10)]

mapping = map(myfunc, ls)

print(list(mapping))
````

### Sorted

The sorted() function returns a sorted list of the specified iterable object.


```
sorted(iterable, key=key, reverse=reverse)
```
Note: You cannot sort a list that contains BOTH string values AND numeric values.


````python
ls = (1,5,2,8,3,6)

sorted_tuple = sorted(ls, reverse=True)
print(sorted_tuple)

````

````python
ls = ("apple", "banana", "cat")

sorted_list = sorted(ls, key=len) #returns a sorted list 

print(sorted_list)

sorted_list.append("mango")
print(sorted_list)
````

Sort the list by the number closest to 10:

````python
def myfunc(n):
    return abs(10-n)


# ls = [i for i in range(10)]

ls = (5, 3, 1, 11, 2, 12, 17)

sorted_arr = sorted(ls)
print(sorted_arr)


sorted_fnc = sorted(ls, key=myfunc)

print(sorted_fnc)

````
