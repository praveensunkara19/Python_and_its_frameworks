#### Set

* Sets are used to store multiple items in a single variable.
* A set is a collection which is unordered, unchangeable*, and unindexed. Not allow duplicate values.
* Once a set is created, you cannot change its items, but you can remove items and add new items.(add/remove)


```python
a = {2,4,5}
b = {4,5,6}

print(a & b)
print(a.intersection(b))


print(a | b)
print(a.union(b))

print(a ^ b) #symmetric diff
```

```python
thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
a.add(7)
a.remove(2)
a
```

```python
a.intersection(b)
```

Note: The values False and 0 are considered the same value in sets, and are treated as duplicates:

```python
thisset = {"apple", "banana", "cherry", False, True, 0}

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}
tropical = {"pineapple", "mango", "papaya"}

thisset.add("zebra")  #adds element

thisset.update(tropical) #concatinates and updates the the set

print(thisset)
```

Add Any Iterable

The object in the update() method does not have to be a set, it can be any iterable object (tuples, lists, dictionaries etc.).

```python
thisset = {"apple", "banana", "cherry"}
mylist = ["kiwi", "orange"]

thisset.update(mylist)  #updates work with any iterable

print(thisset)
```

```python
dictionary = {"a":"apple", "b":"banana", "c":"cat"}
thisset.update(dictionary)
thisset
```

join multiple sets

```python
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

myset = set1.union(set2, set3, set4)
print(myset)
```
