### Match Statement

```
match expression:
  case x:
    code block
  case y:
    code block
  case z:
    code block

````python
day = 4
match day:
  case 1:
    print("Monday")
  case 2:
    print("Tuesday")
  case 3:
    print("Wednesday")
  case 4:
    print("Thursday")
  case 5:
    print("Friday")
  case 6:
    print("Saturday")
  case 7:
    print("Sunday")
````

Default value _

````python
day = 4
match day:
  case 6:
    print("Today is Saturday")
  case 7:
    print("Today is Sunday")
  case _:
    print("Looking forward to the Weekend")
````

Combine Values

````python
day = 7
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
````

If Satement as Guard

````python
month = 5
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5 if month == 4:
    print("A weekday in April")

  case 1 | 2 | 3 | 4 | 5 if month == 5:
    print("A weekday in May")
    
  case _:
    print("No match")
````

### While Loop

* break -> to stop the loop
* continue -> stop the current execution, proceed the next iteration

````python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
````

````python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
  
````

Note: The else block will NOT be executed if the loop is stopped by a break statement.



````python
i = 1
while i < 6:
  print(i)
  i += 1
#   if i == 6:
#     break
else:
  print("i is no longer less than 6")
````

### For looops

````python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
````

````python
a = "banana"
for x in a:
  print(x)
````

````python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
````

````python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
````
