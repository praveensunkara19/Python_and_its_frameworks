#### Arguments
Information can be passed into functions as arguments.

Parameters vs Arguments
```
From a function's perspective:
A parameter is the variable listed inside the parentheses in the function definition.
An argument is the actual value that is sent to the function when it is called.


````python
def my_function(name): # name is a parameter
  print("Hello", name)

my_function("Emil") # "Emil" is an argument
````

Defualt parameters values

````python
def my_function(name = "friend"):
  print("Hello", name)

my_function("Emil")
my_function("Tobias")
my_function()
my_function("Linus")
````

````python
def my_function(animal, **name):
  print("I have a", animal)
  print("My", animal + "'s name is", name)

my_function("Buddy", name="dog")
````

Arguments and Key Word Aruguments

```
* args (Arbitrary Positional Arguments) -> collects multiple positional arguments into a tuple.
* kwargs` (Arbitrary Keyword Arguments) -> collects multiple keyword arguments into a dictionary

Positional-Only Arguments
* specify positional-only arguments, add , / after the arguments:

````python
def my_function(name, age,/):
  print("Hello", name)
  print("age", age)

# my_function("Emil", age=25) #-->restricting the kwargs
my_function("Emil", 25)
````

Keyword-Only Arguments

````python
def my_func(*, name, age):
    print("name", name)
    print("age", age)

# my_func('praveen', 23) #--> restricting the positional args
my_func(name='praveen', age=23)
````

#### *args & **kwargs

````python
def greet(first, name, *arg,**kwargs):       #--> parameters
    print(type(arg))
    print(type(kwargs))
    return(f"{first}, {name}, {arg}, {kwargs}")
    

# greet("praveen", "sunkara")  # ---> positional auguments 
# greet("nani", "sunkara", "there")
greet("nani", "sunkara", "hello", fname="sunkara",lname="sridevi")
````

```
Arbitrary Arguments - *args
If you do not know how many arguments will be passed into your function, add a * before the parameter name.

This way, the function will receive a tuple of arguments and can access the items accordingly:

````python
def my_function(*kids):
  print("The youngest child is " + kids[2]) #-> collects list of arguments.

my_function("Emil", "Tobias", "Linus")
````

Arbitary Arguments -> *args
* The *args parameter allows a function to accept any number of positional arguments.

````python
def my_function(greeting, *names):
  print(type(names))
  for name in names:
    print(greeting, name)

my_function("Hello", "Emil", "Tobias", "Linus")
````

Arbitrary Keyword Arguments - **kwargs
* The **kwargs parameter allows a function to accept any number of keyword arguments.

````python
def my_function(**kid):
  print(type(kid))
  print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
````

Unpacking Arguments

````python
# using * to unpack the list into arguments

def my_function(a, b, c):
  return a + b + c

numbers = [1, 2, 3]
result = my_function(*numbers) # Same as: my_function(1, 2, 3)
print(result)
````

````python
# using ** to unpack the dict into auguments

def my_func(a, b, c):
    print('name_a', a)
    print('name_b', b)
    print('name_c', c)

args = {'b':"praveen_b", 'a':'praveen_a', 'c':"praveen_c"}
my_func(**args)
````

#### Scope

Global variables

````python
# global x
x = 10

def my_func():

    global y

    y = x+20
    print(y)

my_func()  #if we define the global variable inside the func, then the func need to call firt.

print(y)

````

````python
x = 300

def myfunc():
  global x
  x = 200

myfunc()

print(x)
````

non local 
* If you use the nonlocal keyword, the variable will belong to the outer function:


````python
def my_func1():

    x = "jane"

    def my_func2():
        nonlocal x 
        x = "megana"
    my_func2()

    print(x)

my_func1()
````

````python
x = "global"

def outer():
  x = "enclosing"
  def inner():
    x = "local"
    print("Inner:", x)
  inner()
  print("Outer:", x)

outer()
print("Global:", x)
````

Arguments in the Decorators

````python
def changecase(func):

    def inner(x):
        return func(x).upper()
    
    return inner


@changecase
def my_func(name):
    return f"hello {name}"

print(my_func("praveen"))
````

#### *args and **kwargs
* to control the arguments passed to the arguments

````python
def decorator(func):

    def inner(*args,**kwargs):
        return func(*args, **kwargs).upper()
    
    return inner


@decorator
def myfunc(name):
    return "hello"+name

myfunc("pravEEN")
````

Decorator With Arguments

````python
def changecase(n):

    def changecase(func):
        def inner(x):
            if n == 1:
                return func(x).upper()
            else:
                return func(x).lower()
        return inner
    
    return changecase



@changecase(1)
def myfunc1(name):
    return "Hello"+name

@changecase(2)
def myfunc2(name):
    return "Hello"+name

print(myfunc1("PraveeN"))
print(myfunc2("pRAVEEn"))
````

Multiple Decorators

````python
def upperDecorator(func):
    def inner(x):
        return func(x).upper()
    return inner


def greetDecorator(func):
    def inner(x):
        return "hello "+ func(x) + " greetings of the day!"
    return inner


@upperDecorator
@greetDecorator
def myfunc(name):
    return name

print(myfunc("nani"))
 
````

Preserving Function MetaData

````python
def myfunction():
  return "Have a great day!"

print(myfunction.__name__)
````

````python
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Have a great day!"

print(myfunction.__name__)
````

To fix this, Python has a built-in function called functools.wraps that can be used to preserve the original function's name and docstring.

````python
import functools

def changecase(func):
  @functools.wraps(func)
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Have a great day!"

print(myfunction.__name__)
````

#### Lambda Functions
A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

```
lamda arguments : expression

````python
x = lambda a : a + 10

print(x(5))
````

````python
x = lambda a,b : a * b
print(x(3,4))
````

````python
def myfunc(n):
    return a*n

double = myfunc(3)
double(2)

````

to handle this lamda a:a*n

````python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
````

Use lambda functions when an anonymous function is required for a short period of time.

````python
nums = [1,2,3,4,5]

sq = [i**2 for i in nums]
sq
````

#### Lamda functions with map(), filter(), sorted()

Using Lambda with map()

The map() function applies a function to every item in an iterable:

````python
double = list(map(lambda x: x*2, nums))
double
````

````python
sq = list(map(lambda a:a*a, nums))

print(sq)
````

Using Lambda with filter()

The filter() function creates a list of items for which a function returns True:

````python
odds = list(filter(lambda x: x%2 != 0, nums))

odds
````

Using Lambda with sorted()

The sorted() function can use a lambda as a key for custom sorting:

````python
students = [("Emil", 25), ("Tobias", 22), ("Linus", 28)]  #dict items

sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
````

````python
student = {"a":25, "c":20, "b":27}

print(student.items())

sorted_keys = sorted(student.items())
print(sorted_keys)

sorted_values = sorted(student.items(), key=lambda x:x[1], reverse=True)
print(sorted_values)
````

````python
words = ["apple", "pie", "banana", "cherry"]
sorted_words = sorted(words, key=lambda x : len(x))
print(sorted_words)
````
