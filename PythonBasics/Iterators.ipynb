Iterators

* An iterator is an object that contains a countable number of values.
* That you can iterate on it, traverse through the all values.

List,Tuple,Dictionaries, set all are iterable objects.

```python
ls = [1,2,4,5,6] #iterable
ls[0]
```

```python
my_list = iter(ls)  #iterator
next(my_list)
next(my_list)
```

Even Strings are iterable objects,  we can return iterator on it.

```python
mystr = "banana"
myit = iter(mystr)

print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
print(next(myit))
```

Using for loop to iterate on itarable objects.

```python
mytuple = ("apple", "banana", "cherry")

for x in mytuple:
  print(x)
```

```python
class MyNumber():
    def __iter__(self): #intializing a = 1 like __init__()
        self.a = 1
        return self
    
    def __next__(self):
        x = self.a
        self.a +=1
        return x
    
myClass = MyNumber()
myiter = iter(myClass)

print(next(myiter))
print(next(myiter))

```
