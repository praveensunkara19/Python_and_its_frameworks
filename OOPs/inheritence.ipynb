```python
class Parent:

    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname

    def getName(self):
        return self.fname, self.lname
    
obj = Parent("Praveen", "Sunkara")

obj.getName()
```

```python
class Child(Parent):
    # pass

    def __init__(self,fname, lname, phone):   #overrides the parents __init__() method, so we need to define the var again
        super().__init__(fname, lname)     #passing the values to the super class
        self.phone  = phone
        self.gradyear = 2019

        
    
    def getDetails(self):
        return self.fname, self.lname, self.phone, self.gradyear
    
obj2 = Child("Anand", "kumar", 87)

print(obj2.getDetails())

print(obj2.getName())

    
```
