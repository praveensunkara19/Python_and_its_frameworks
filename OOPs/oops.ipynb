## Inheritance

```python
# parent class

class Person:
 
    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height

    def __str__(self):
        return "name: {}, age: {}, height: {}".format(self.name, self.age, self.height)

        
```

```python
class Worker(Person):

    """accessing from the parent class Person"""
    def __init__(self, name, age, height, salary):
        super().__init__(name, age, height)    #super() allows you can access the parent variables
        self.salary = salary


    """overriding the parent method"""
    def __str__(self):
        return super().__str__() + " salary: {}".format(self.salary)
    
    def year_salary(self):
        return self.salary * 12
    
    
work_obj = Worker("Ram", 25, 168, 5000)

print(work_obj.year_salary())

print(work_obj)
```

## Operator Overloading

```python
class Vector:
    """constructor"""
    def __init__(self, x, y):
        self.x = x
        self.y = y

    """adding two vectors returns the vector"""
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    """subtracts the two vectors returns the vector"""
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    
    """representing the given vectors"""
    def __str__(self):
        return "X: {}, Y: {}".format(self.x, self.y)
    

v1 = Vector(2,3)
v2 = Vector(4,5)

print(v1)
print(v2)

v_add = v1 + v2
v_sub = v1 - v2

print(f"adding [{v1}] and [{v2}]: {v_add}")
print(f"subtracting {v1} and {v2}: {v_sub}")

```

## Sychronization

* Locking enables to use the the data for the functions one by one Ex: booking seat, when user selects and way to payments, while other cant book the same utill the lock is released(i.e booked or cancelled)

```python
import threading
import time


'''
This function threads use the global vairable simultaniously which cause the disturbance it the code run.
'''
x = 8192

def double():
    global x
    while x < 16384:
        x *=2
        print(x)
        time.sleep(1)
    print("reached maxium")

def halve():
    global x
    while x > 1:
        x /= 2
        print(x)
        time.sleep(1)
    print("reached minimum")

t1 = threading.Thread(target=halve)
t2 = threading.Thread(target=double)

t1.start()
t2.start()
```

```python
import threading
import time

'''
Handling with the locks
This enable to run only the selected threads and until the lock is released
'''
x = 8192
lock = threading.Lock()

def double():
    global x, lock
    lock.acquire()  #initializing lock
    print("double thread locked")
    while x < 16384:
        x *=2
        print(x)
        time.sleep(1)
    print("reached maxium")
    lock.release()
    print("thread realeased")

def halve():
    global x, lock

    lock.acquire()
    print("halve thread locked")
    while x > 1:
        x /= 2
        print(x)
        time.sleep(1)
    print("reached minimum")
    lock.release()
    print("thread released")

t1 = threading.Thread(target=halve)
t2 = threading.Thread(target=double)

t1.start()
t2.start()
```

### Semaphores

```python
import threading
import random

semaphore = threading.BoundedSemaphore(value=5)

x = 100

def access(thread_num):
    global x

    print("{} thread tring for the access".format(thread_num))
    semaphore.acquire()

    print("{} this thread got access!".format(thread_num))
    if x > 0:
        print(f"{thread_num} took this Mangoes: ", 20)
        x -= 20
        time.sleep(2)
    else:
        print("Sorry!, no more Mangos left!")
    print("{} this thread it going back".format(thread_num))
    semaphore.release()


for thread_num in range(1,20):
    thread = threading.Thread(target=access, args=(thread_num,))
    thread.start()
    time.sleep(5)


```

## Events 

* Events are used to get the instructions from the user to get initialize the code in the threads!

```python
import threading

event = threading.Event()

def myfunction():
    print("Waiting for the Event to trigger:")
    event.wait()   # this waits for the instruction by the user!
    print("Event is trigger: processing XYZ...")

t1 = threading.Thread(target=myfunction)
t1.start()

x = input("Do you want to process the func: (y/n):")

if x == "y":
    event.set()
else:
    print("user choosed no")

```

## Daemon Threads

* Daemon Threads are used to get the data synchronous from the other defied files/threads. 

```python
import threading
import time

path = "text.txt"
text = ""
def readFile():
    global text,path
    while True:
        with open(path, 'r') as f:
            text = f.read()
        time.sleep(1)

def printLoop():
    for x in range(30):
        print(text)
        time.sleep(1)


t1 = threading.Thread(target=readFile, daemon=True)
t2 = threading.Thread(target=printLoop)

t1.start()
t2.start()
```
