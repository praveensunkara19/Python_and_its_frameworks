### Polymorphism (Many Ways) - doing a task in many ways

* ex: len() for list,dict,tuple

````python
class Boat:

    def __init__(self, name, brand):
        self.name = name
        self.brand = brand
    def move(self):
        print("Sail")

class Car:

    def __init__(self, name, brand):
        self.name = name
        self.brand = brand
    def move(self):
        print("Drive")

class Plane:

    def __init__(self, name, brand):
        self.name = name
        self.brand = brand
    def move(self):
        print("Fly")


boat1 = Boat("Ibiza", "Touring 20")
car1 = Car("Ford", "Mustang")
plane1 = Plane("Boeing", "747")

for i in (boat1, car1, plane1):
    i.move()
        
````

````python
class Parent:

    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname

    def get_details(self):
        return self.fname, self.lname


class Child(Parent):
    def __init__(self, fname, lname, age):
        super().__init__(fname, lname)
        self.age = age
    
    def get_details(self):
        return self.fname, self.lname, self.age
    
    

obj0 = Parent("Das", "sunkara")
obj1 = Child("praveen", "sunkara", 23)

print(obj0.get_details())
print(obj1.get_details())



````

Static methods

````python
class A:
    @staticmethod
    def add(x, y):
        return x + y
    


A.add(2,3)
````

Class methods
```
Use when:
* You need to modify class-level data
* Alternative constructors.

````python
class A:
    count = 0

    def __init__(self):
        A.count += 1

    @classmethod
    def how_many(cls):
        return cls.count


A()
A()
print(A().how_many())
print(A().how_many())
````
