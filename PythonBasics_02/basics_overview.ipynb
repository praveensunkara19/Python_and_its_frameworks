### Recursive functions

```python
#1235

sum = 0
num = 1234

def reverse(num,sum):
    if num == 0:
        return sum
        
    else:
        x = num%10
        sum = sum*10+x
        
    return reverse(num//10,sum)

reverse(num,sum)
    
    
```
