#### Identity Operators

* is, is not

* not compares the values, compares the memory allocation of the objects.
```
is - Checks if both variables point to the same object in memory
== - Checks if the values of both variables are equal


````python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is y)  #compares the memory location of the objects reference.
print(x is z)
print(x == y)  #compares the values of the object.

print(x is not y) # is not operator returns True if both variables do not point to the same object:
````

#### Membership Operators
* in, not in

Membership operators are used to test if a sequence is presented in an object:

````python
x = 'abced'

print('a' in x)
print('f' not in x)
````

````python
fruits = ["apple", "banana", "cherry"]

print("pineapple" not in fruits)
````

#### Bitwise Operators
Bitwise operators are used to compare (binary) numbers:
```
&   AND            1 & 1 -> 1
|   OR             1 | 0 -> 1
^   XOR            0 ^ 1 -> 1
`   NOT            1 -> 0, 0 -> 1
<<  LEFT SHIFT     adds bits to the right
>> RIGHT SHIFT     removes the bits from the right

````python
x = 8; y = 7
print(f'bits:{x}', bin(x))
print(f'bits:{y}', bin(y))

print(f'{x}&{y}:',x&y, bin(x&y))
print(f'{x}|{y}:',x|y, bin(x|y))
print(f'{x}^{y}:',x^y, bin(x^y))
print(f'{x}<<2', x<<2, bin(x<<2))
print(f'{x}>>2', x>>2, bin(x>>2))
````
